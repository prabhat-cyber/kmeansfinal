```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Replace 'your_csv_file.csv' with the actual path to your CSV file
df = pd.read_csv("C:\\Users\\91892\\OneDrive\\Desktop\\OnlineRetail.csv", encoding='latin1')
```

```python
# Handling missing values by dropping rows with missing values
df.dropna(axis=0, inplace=True)
```

```python
# Removing duplicate entries in the dataset
df.drop_duplicates(inplace=True)
```

```python
# Convert CustomerID from float to int
df['CustomerID'] = df['CustomerID'].astype(int)

# Display datatypes after conversion
print(df.dtypes)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Standardization
scaler = StandardScaler()
numerical_columns = ['Quantity', 'UnitPrice', 'CustomerID']
df_standardized = df.copy()
df_standardized[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Display standardized dataframe
print(df_standardized.head())
```

```python
# Min-Max Scaling (Normalization)
scaler = MinMaxScaler()
df_normalized = df.copy()
df_normalized[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
# Display normalized dataframe
print(df_normalized.head())
```

```python
# Encoding categorical variables
df_encoded = pd.get_dummies(df_normalized, columns=['Country', 'Description', 'StockCode'], drop_first=True)
print(df_encoded.head())
```

```python
df_fe = df_encoded.copy()

# Creating new features based on existing ones
# Create a feature for total sales
df_fe['TotalSales'] = df_fe['Quantity'] * df_fe['UnitPrice']

# Create a feature for the day of the week (Monday = 0, Sunday = 6)
df_fe['DayOfWeek'] = df_fe['InvoiceDate'].dt.dayofweek

# Create a feature for the hour of the day
df_fe['HourOfDay'] = df_fe['InvoiceDate'].dt.hour

print(df_fe.head())
```

```python
# Calculate correlation matrix
correlation_matrix = df_fe[['Quantity', 'UnitPrice', 'CustomerID', 'TotalSales', 'DayOfWeek', 'HourOfDay']].corr()

# Visualize using a heatmap to identify highly correlated features
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Assuming df_encoded is your DataFrame after encoding categorical variables

# Drop non-numerical columns like 'InvoiceDate' if it's causing issues
# Example: df_encoded.drop(columns=['InvoiceDate'], inplace=True)

# Select numerical columns for clustering
numerical_cols = df_encoded.select_dtypes(include=['number'])

# Determine the optimal number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(numerical_cols)  # Use only numerical columns for clustering
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Min-Max Scaling (Normalization)
scaler = MinMaxScaler()
df_normalized = df.copy()
df_normalized[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
nan_columns = df_standardized.columns[df_standardized.isna().any()].tolist()
print("Columns with NaN values:", nan_columns)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
df_standardized[numerical_columns] = imputer.fit_transform(df_standardized[numerical_columns])
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(df_standardized[numerical_columns])
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Elbow Method to find the optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(df_standardized[numerical_columns])
    wcss.append(kmeans.inertia_)

# Plot the WCSS against number of clusters
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

```python
# Based on the Elbow Method, let's choose the optimal number of clusters (e.g., 3)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42)
kmeans.fit(df_standardized[numerical_columns])

# Predicting clusters
df['Cluster'] = kmeans.labels_
```

```python
# Plotting clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Quantity', y='UnitPrice', hue='Cluster', data=df, palette='viridis')
plt.title('Customer Segmentation')
plt.xlabel('Quantity')
plt.ylabel('Unit Price')
plt.legend(title='Cluster')
plt.show()
```

```python
# Adding cluster labels to original dataframe
df['Cluster'] = kmeans.labels_
```

```python
# Summary statistics for each cluster
cluster_summary = df.groupby('Cluster')[['Quantity', 'UnitPrice', 'CustomerID']].describe()

# Example of accessing statistics
print(cluster_summary)
```
